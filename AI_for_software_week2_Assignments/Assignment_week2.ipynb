```python
import pandas as pd

# Load your file
df = pd.read_csv("Production_Crops_E_Asia.csv")

# Melt the dataset: keep Area + Item, and convert year columns into rows
df_melted = df.melt(
    id_vars=["Area", "Item"],
    value_vars=[col for col in df.columns if col.startswith("Y") and not col.endswith("F")],
    var_name="Year",
    value_name="Production"
)

# Clean the Year column (remove the "Y" prefix and convert to integer)
df_melted["Year"] = df_melted["Year"].str[1:].astype(int)

# Drop rows where Production is missing
df_melted = df_melted.dropna(subset=["Production"])

# Preview the cleaned data
df_melted.head()
```

```python
# Filter for one country and one crop
crop_data = df_melted[(df_melted["Area"] == "China") & (df_melted["Item"] == "Wheat")]

# Sort by Year
crop_data = crop_data.sort_values("Year")

# Features (X) = Year, Target (y) = Production
X = crop_data[["Year"]]
y = crop_data["Production"]

crop_data.head()
```

```python
# Check unique country names
print("Areas:", df_melted["Area"].unique()[:20])

# Check unique crop names
print("Items:", df_melted["Item"].unique()[:20])
```

```python
# Filter for one country and one crop (Maize in China, mainland)
crop_data = df_melted[(df_melted["Area"] == "China, mainland") & (df_melted["Item"] == "Maize")]

# Sort by Year
crop_data = crop_data.sort_values("Year")

# Features (X) = Year, Target (y) = Production
X = crop_data[["Year"]]
y = crop_data["Production"]

crop_data.head()
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Train model
model = LinearRegression()
model.fit(X, y)

# Predict production for past + next 10 years
future_years = pd.DataFrame({"Year": list(range(1961, 2030))})
predictions = model.predict(future_years)

# Plot actual vs predicted
plt.figure(figsize=(10,5))
plt.scatter(X, y, label="Actual Data")
plt.plot(future_years["Year"], predictions, color="red", label="Predicted Trend")
plt.xlabel("Year")
plt.ylabel("Production")
plt.title("Maize Production in China, mainland (Prediction)")
plt.legend()
plt.show()
```

```python
# Select multiple crops for one country
crops = ["Maize", "Barley", "Potatoes"]

for crop in crops:
    crop_data = df_melted[(df_melted["Area"] == "China, mainland") & (df_melted["Item"] == crop)]
    crop_data = crop_data.sort_values("Year")

    X = crop_data[["Year"]]
    y = crop_data["Production"]

    model = LinearRegression()
    model.fit(X, y)

    future_years = pd.DataFrame({"Year": list(range(1961, 2030))})
    predictions = model.predict(future_years)

    plt.figure(figsize=(10,5))
    plt.scatter(X, y, label="Actual Data")
    plt.plot(future_years["Year"], predictions, color="red", label="Predicted Trend")
    plt.xlabel("Year")
    plt.ylabel("Production")
    plt.title(f"{crop} Production in China, mainland (Prediction)")
    plt.legend()
    plt.show()
```

```python
from sklearn.metrics import mean_absolute_error

for crop in crops:
    crop_data = df_melted[(df_melted["Area"] == "China, mainland") & (df_melted["Item"] == crop)]
    crop_data = crop_data.sort_values("Year")

    X = crop_data[["Year"]]
    y = crop_data["Production"]

    model = LinearRegression()
    model.fit(X, y)
    y_pred = model.predict(X)

    mae = mean_absolute_error(y, y_pred)
    print(f"{crop} - Mean Absolute Error: {mae:.2f}")
```
